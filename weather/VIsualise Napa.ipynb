```python
import numpy as np
import pandas as pd
import calendar

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('napa.txt', header=None).replace(r'\N', np.NaN)
df.columns = [
    'decade',
    'stationid',
    'month',
    'max_temp',
    'min_temp',
    'avg_temp',
    'max_humid',
    'min_humid',
    'avg_humid',
    'max_ws',
    'min_ws',
    'avg_ws',
    'max_p1',
    'min_p1',
    'avg_p1',
    'max_p6',
    'min_p6',
    'avg_p6'
]
df.head()
```

### Average Temperature

```python
# color cycle
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

plt.figure(figsize=(15,10))

i = 0
for cur_decade in df['decade'].unique():
    if cur_decade > 198:  # only data from the 90s are reliable
        sub_df = df[df['decade'] == cur_decade]
        sorted_df = sub_df[['month', 'avg_temp', 'max_temp', 'min_temp']].astype(float).sort_values('month')
        sorted_df['month'] = sorted_df['month'].apply(lambda x: calendar.month_name[int(x)])
        
        plt.plot(sorted_df['month'], sorted_df['avg_temp']/10, 
                 label=str(cur_decade * 10) + "s avg", c=colors[i])
        plt.plot(sorted_df['month'], sorted_df['max_temp']/10, 
                 label=str(cur_decade * 10) + "s max", c=colors[i], linestyle='--', alpha=0.5)
        plt.plot(sorted_df['month'], sorted_df['min_temp']/10, 
                 label=str(cur_decade * 10) + "s min", c=colors[i], linestyle='--', alpha=0.5)
        i += 1
plt.legend()
plt.show()
```
